```python
import littlefs
cfg = littlefs.LFSConfig(
    block_size=4096,       # Must match flash erase size (typically 4 KB)
    block_count=8,         # 32 KB / 4 KB = 8 blocks
    read_size=16,          # Match Zephyr's `read-size`
    prog_size=16,          # Match Zephyr's `prog-size`
    cache_size=128,        # Match Zephyr's `cache-size`
    lookahead_size=16,     # Match Zephyr's `lookahead-size`
    block_cycles=256,      # Wear-leveling cycles
)

# Create a new filesystem image
fs = littlefs.LFSFilesystem(cfg)
littlefs.
print(type(fs))

#fs.mkfs()  # Format the image

# Mount and add test files
fs.mount()
fs.mkdir("/lfs")          # Create mount point (match Zephyr's `mount-point`)
with fs.open("/lfs/test.txt", "w") as f:
    f.write("Hello from Python LittleFS!")  # Write a test file

# Dump filesystem to a binary file
with open("littlefs.img", "wb") as f:
    f.write(fs.context.buffer)

print("LittleFS image created: littlefs.img")
```
